```python
# Calculate all sales data first
category_sales = df.groupby("Category")["Sales"].sum()
region_sales = df.groupby("Region")["Sales"].sum()
segment_sales = df.groupby("Segment")["Sales"].sum()
sub_sales = df.groupby("Sub_Category")["Sales"].sum().sort_values(ascending=False)
df["Order_Date"] = pd.to_datetime(df["Order_Date"])
df["Year"] = df["Order_Date"].dt.year
year_sales = df.groupby("Year")["Sales"].sum()

print("✅ All sales data ready!")
```

```python
fig, axs = plt.subplots(2, 3, figsize=(18, 10))

axs[0,0].bar(category_sales.index, category_sales.values)
axs[0,0].set_title("Sales by Category")

axs[0,1].bar(region_sales.index, region_sales.values)
axs[0,1].set_title("Sales by Region")

axs[0,2].pie(segment_sales.values, labels=segment_sales.index, autopct="%1.1f%%")
axs[0,2].set_title("Sales by Segment")

axs[1,0].bar(sub_sales.index[:8], sub_sales.values[:8])
axs[1,0].set_title("Top Sub Categories")
axs[1,0].tick_params(axis="x", rotation=45)

axs[1,1].plot(year_sales.index, year_sales.values)
axs[1,1].set_title("Yearly Sales Trend")

axs[1,2].scatter(df["Sales"], df["Profit"])
axs[1,2].set_title("Sales vs Profit")

fig.suptitle("Superstore Sales Dashboard", fontsize=18)
plt.tight_layout()
plt.savefig("superstore_dashboard.png", bbox_inches="tight", dpi=150)
plt.show()
print("✅ Superstore Dashboard saved!")
```
